각도 계산 코드 (x, y축만 고려)

````python

# 1. 필요한 라이브러리 설치
!pip install -q mediapipe

# 2. 라이브러리 임포트
import cv2
import os
import math
import mediapipe as mp
from mediapipe.tasks import python
from mediapipe.tasks.python import vision
from google.colab.patches import cv2_imshow
from google.colab import drive

# 3. 구글 드라이브 연결
drive.mount('/content/drive')

# 4. MediaPipe Pose Landmarker 모델 다운로드 및 설정
if not os.path.exists('pose_landmarker.task'):
    os.system('wget -q -O pose_landmarker.task https://storage.googleapis.com/mediapipe-models/pose_landmarker/pose_landmarker_heavy/float16/1/pose_landmarker_heavy.task')

base_options = python.BaseOptions(model_asset_path='pose_landmarker.task')
options = vision.PoseLandmarkerOptions(
    base_options=base_options,
    output_segmentation_masks=False)
pose_detector = vision.PoseLandmarker.create_from_options(options)

# 각도 계산 함수 (수평선 기준)
def calculate_angle(p1, p2):
    # p1: C7 (기준점), p2: 귀
    # 수평선과 p1-p2 선분이 이루는 각도 계산
    dy = p1[1] - p2[1]
    dx = p2[0] - p1[0]
    angle = math.degrees(math.atan2(dy, dx))
    # 예각으로 변환 (필요시)
    if angle < 0:
        angle += 360
    if angle > 90 and angle <= 180:
        angle = 180 - angle
    elif angle > 180 and angle <= 270:
        angle = angle - 180
    elif angle > 270:
        angle = 360 - angle
    return angle

# 5. 이미지 처리 및 좌표 추출 메인 함수
def process_pose_image(img_path):
    img = cv2.imread(img_path)
    if img is None:
        return print(f"\u274c 파일을 찾을 수 없습니다. 경로를 확인해주세요: {img_path}")

    h, w, _ = img.shape
    # BGR을 RGB로 변환
    mp_image = mp.Image(image_format=mp.ImageFormat.SRGB, data=cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

    # 포즈 랜드마크 검출
    results = pose_detector.detect(mp_image)

    if results.pose_landmarks:
        landmarks = results.pose_landmarks[0]

        # 1. 얼굴 중심 (코 랜드마크: 인덱스 0)
        nx, ny = int(landmarks[0].x * w), int(landmarks[0].y * h)

        # 2. 왼쪽 어깨 (랜드마크: 인덱스 11)
        lsx, lsy = int(landmarks[11].x * w), int(landmarks[11].y * h)

        # 3. 오른쪽 어깨 (랜드마크: 인덱스 12)
        rsx, rsy = int(landmarks[12].x * w), int(landmarks[12].y * h)

        # 4. 왼쪽 귀 (랜드마크: 인덱스 7), 오른쪽 귀 (랜드마크: 인덱스 8)
        # 측면 사진인 경우 보이는 귀를 선택하는 로직이 필요할 수 있으나, 여기서는 평균 또는 보이는 귀 사용
        lex, ley = int(landmarks[7].x * w), int(landmarks[7].y * h)
        rex, rey = int(landmarks[8].x * w), int(landmarks[8].y * h)

        # C7 위치 근사 (양쪽 어깨의 중간 지점)
        c7x, c7y = int((lsx + rsx) / 2), int((lsy + rsy) / 2)

        # 왼쪽 귀를 기준으로 CVA 각도 계산 (옆모습 기준에 따라 오른쪽 귀 사용 가능)
        ear_x, ear_y = lex, ley # 임시로 왼쪽 귀 사용
        cva_angle = calculate_angle((c7x, c7y), (ear_x, ear_y))

        # 텍스트로 좌표 출력
        print(f"\u2705 얼굴(코) 중심 좌표: X={nx}, Y={ny}")
        print(f"\u2705 왼쪽 어깨 좌표: X={lsx}, Y={lsy}")
        print(f"\u2705 오른쪽 어깨 좌표: X={rsx}, Y={rsy}")
        print(f"\u2705 C7(근사) 좌표: X={c7x}, Y={c7y}")
        print(f"\u2705 귀 좌표: X={ear_x}, Y={ear_y}")
        print(f"\ud83d\udcd0 CVA 각도: {cva_angle:.2f}도")

        # 이미지에 좌표 포인트 및 텍스트 표시
        cv2.circle(img, (nx, ny), 6, (0, 0, 255), -1)
        cv2.circle(img, (lsx, lsy), 6, (0, 255, 0), -1)
        cv2.circle(img, (rsx, rsy), 6, (255, 0, 0), -1)

        # C7 및 귀 표시
        cv2.circle(img, (c7x, c7y), 6, (0, 255, 255), -1)
        cv2.circle(img, (ear_x, ear_y), 6, (255, 255, 0), -1)

        # C7에서 귀를 잇는 선과 수평선 그리기
        cv2.line(img, (c7x, c7y), (ear_x, ear_y), (0, 255, 255), 2)
        cv2.line(img, (c7x, c7y), (c7x + 100, c7y), (255, 255, 255), 2) # 수평선

        # 텍스트 표시
        cv2.putText(img, f"CVA: {cva_angle:.2f} deg", (c7x + 15, c7y - 15), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 255), 2)

    else:
        print("\u26a0\ufe0f 사람의 포즈를 찾을 수 없습니다.")

    cv2_imshow(img)

import glob

photo_folder = '/content/drive/MyDrive/dl_프로젝트/new_total_data'

all_images = sorted(glob.glob(os.path.join(photo_folder, '**/*.jpg'), recursive=True))

print(f"총 {len(all_images)}장 발견\n")

for i, img_path in enumerate(all_images):
    print(f"\n--- [{i+1}/{len(all_images)}] {os.path.basename(img_path)} ---")
    process_pose_image(img_path)

````

````python
# ============================================================
# Step 2: CVA → GT 회전행렬 변환 + JSON 저장
# (이전 셀에서 pose_detector, calculate_angle_2d가 이미 실행된 상태)
# ============================================================

import json
import math
import numpy as np
import glob

def calculate_angle_2d(p1, p2):
    dy = p1[1] - p2[1]
    dx = abs(p2[0] - p1[0])
    angle = math.degrees(math.atan2(dy, dx))
    return angle
# ─────────────────────────────────────
# 설정값
# ─────────────────────────────────────
REFERENCE_CVA = 55.0   # 정상 자세의 기준 CVA (논문: 53°~55°)
photo_folder = '/content/drive/MyDrive/dl_프로젝트/new_total_data'
output_json = '/content/drive/MyDrive/dl_프로젝트/final_GT.json'

# ─────────────────────────────────────
# 함수 1: CVA → pitch 변환
# ─────────────────────────────────────
def cva_to_pitch(cva_angle):
    """
    CVA 55° → pitch 0° (정상)
    CVA 35° → pitch 20° (거북목)
    """
    pitch_deg = REFERENCE_CVA - cva_angle
    pitch_rad = pitch_deg * math.pi / 180.0
    return pitch_rad, pitch_deg

# ─────────────────────────────────────
# 함수 2: 오일러각 → 회전행렬 (utils.py의 get_R과 동일)
# ─────────────────────────────────────
def get_R(pitch, yaw, roll):
    Rx = np.array([[1, 0, 0],
                   [0, np.cos(pitch), -np.sin(pitch)],
                   [0, np.sin(pitch), np.cos(pitch)]])
    Ry = np.array([[np.cos(yaw), 0, np.sin(yaw)],
                   [0, 1, 0],
                   [-np.sin(yaw), 0, np.cos(yaw)]])
    Rz = np.array([[np.cos(roll), -np.sin(roll), 0],
                   [np.sin(roll), np.cos(roll), 0],
                   [0, 0, 1]])
    R = Rz.dot(Ry.dot(Rx))
    return R

# ─────────────────────────────────────
# 함수 3: 측면 사진 1장에서 CVA 추출 (값만 반환)
# ─────────────────────────────────────
def extract_cva(img_path):
    img = cv2.imread(img_path)
    if img is None:
        return None
    h, w, _ = img.shape
    mp_image = mp.Image(
        image_format=mp.ImageFormat.SRGB,
        data=cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    results = pose_detector.detect(mp_image)

    if results.pose_landmarks:
        landmarks = results.pose_landmarks[0]
        ls = (landmarks[11].x * w, landmarks[11].y * h)
        rs = (landmarks[12].x * w, landmarks[12].y * h)
        ear = (landmarks[7].x * w, landmarks[7].y * h)
        c7 = ((ls[0] + rs[0]) / 2, (ls[1] + rs[1]) / 2)
        cva = calculate_angle_2d(c7, ear)
        return cva
    return None

# ─────────────────────────────────────
# 실행: 전체 측면 사진 처리
# ─────────────────────────────────────

# 측면 사진 전부 찾기
side_images = sorted(glob.glob(os.path.join(photo_folder, '**/*side*.jpg'), recursive=True))
print(f"📷 측면 사진 {len(side_images)}장 발견\n")

gt_data = []
skip_count = 0

for i, side_path in enumerate(side_images):

    # ① 대응하는 정면 사진 경로 만들기 (side → front)
    folder = os.path.dirname(side_path)
    filename = os.path.basename(side_path).replace('_side_', '_front_')
    front_path = os.path.join(folder, filename)

    has_front = os.path.exists(front_path)

    # ② 측면 사진에서 CVA 추출
    cva = extract_cva(side_path)
    if cva is None:
        print(f"  ⚠️ [{i+1}] 포즈 검출 실패: {os.path.basename(side_path)}")
        skip_count += 1
        continue

    # ③ CVA → pitch 변환
    pitch_rad, pitch_deg = cva_to_pitch(cva)

    # ④ 회전행렬 생성 (정면 촬영이므로 yaw=0, roll=0)
    R = get_R(pitch_rad, 0.0, 0.0)

    # ⑤ 거북목 등급 판정
    if cva > 50:
        grade = "정상"
    elif cva > 30:
        grade = "경미한 거북목"
    else:
        grade = "심한 거북목"

    # ⑥ 저장
    gt_entry = {
        'front_image': front_path if has_front else '정면사진없음',
        'side_image': side_path,
        'cva_angle': round(cva, 2),
        'pitch_deg': round(pitch_deg, 2),
        'pitch_rad': round(pitch_rad, 4),
        'yaw_deg': 0.0,
        'roll_deg': 0.0,
        'grade': grade,
        'rotation_matrix': R.tolist(),
        'has_front': has_front
    }
    gt_data.append(gt_entry)

    front_status = "✅" if has_front else "❌ 정면없음"
    print(f"  [{i+1}] {os.path.basename(side_path)}: "
          f"CVA={cva:.1f}° → pitch={pitch_deg:.1f}° "
          f"[{grade}] {front_status}")

# ─────────────────────────────────────
# JSON 저장
# ─────────────────────────────────────
with open(output_json, 'w', encoding='utf-8') as f:
    json.dump(gt_data, f, indent=2, ensure_ascii=False)

# ─────────────────────────────────────
# 결과 요약
# ─────────────────────────────────────
print(f"\n{'='*55}")
print(f"📊 결과 요약")
print(f"{'='*55}")
print(f"  처리 성공: {len(gt_data)}장")
print(f"  처리 실패: {skip_count}장")
print(f"  정면 paired: {sum(1 for d in gt_data if d['has_front'])}장")
print(f"  정면 없음:   {sum(1 for d in gt_data if not d['has_front'])}장")
print(f"{'='*55}")

# 등급별 분포
grades = [d['grade'] for d in gt_data]
print(f"  정상:         {grades.count('정상')}장")
print(f"  경미한 거북목: {grades.count('경미한 거북목')}장")
print(f"  심한 거북목:   {grades.count('심한 거북목')}장")
print(f"{'='*55}")

# CVA 통계
cvas = [d['cva_angle'] for d in gt_data]
if cvas:
    print(f"  CVA 최소: {min(cvas):.1f}°")
    print(f"  CVA 최대: {max(cvas):.1f}°")
    print(f"  CVA 평균: {np.mean(cvas):.1f}°")
print(f"{'='*55}")
print(f"\n💾 저장 완료: {output_json}")
# ```

# ---

# ### 실행하면 이런 결과가 나옵니다
# ```
# 📷 측면 사진 15장 발견

#   [1] kki_side_10_010.jpg: CVA=66.4° → pitch=-11.4° [정상] ✅
#   [2] kki_side_10_030.jpg: CVA=64.9° → pitch=-9.9°  [정상] ✅
#   [3] kki_side_10_040.jpg: CVA=72.0° → pitch=-17.0° [정상] ✅
#   [4] kki_side_10_051.jpg: CVA=43.2° → pitch=11.8°  [경미한 거북목] ✅
#   ...

# =======================================================
# 📊 결과 요약
# =======================================================
#   처리 성공: 15장
#   처리 실패: 0장
#   정면 paired: 12장
#   정면 없음:   3장
# =======================================================
#   정상:         10장
#   경미한 거북목: 4장
#   심한 거북목:   1장
# =======================================================
#   CVA 최소: 28.5°
#   CVA 최대: 72.0°
#   CVA 평균: 57.3°
# =======================================================

# 💾 저장 완료: /content/drive/MyDrive/dl_프로젝트/gt_annotations.json
# ```

# ---

# ### 주의할 점 하나!

# 파일 이름에서 `side`를 `front`로 바꾸는 방식으로 정면 사진을 찾습니다:
# ```
# 측면: kki_side_10_010.jpg
# 정면: kki_front_10_010.jpg  ← 자동으로 이 이름을 찾음
````

````python
# ============================================================
# Step 3: gt_annotations.json → TokenHPE용 Dataset 만들기
# ============================================================

import json
import numpy as np
import torch
from torch.utils.data.dataset import Dataset
from PIL import Image
from torchvision import transforms

# ─────────────────────────────────────
# 1. Dataset 클래스 (datasets.py에 들어갈 내용)
# ─────────────────────────────────────
class TurtleNeckDataset(Dataset):
    def __init__(self, json_path, transform=None):
        with open(json_path, 'r', encoding='utf-8') as f:
            all_data = json.load(f)

        # 정면 사진이 있는 것만 사용
        self.data = [d for d in all_data if d['has_front']]
        self.transform = transform
        print(f"✅ 데이터셋 로드: {len(self.data)}장 (정면 paired)")

    def __getitem__(self, index):
        entry = self.data[index]

        # ① 정면 사진 로드 (모델 입력!)
        img = Image.open(entry['front_image']).convert('RGB')

        # ② GT 회전행렬 [3,3]
        R = np.array(entry['rotation_matrix'])

        # ③ GT 오일러각 (degree) — Identity 함수용
        cont_labels = torch.FloatTensor([
            entry['pitch_deg'],
            entry['yaw_deg'],
            entry['roll_deg']
        ])

        # ④ 이미지 전처리
        if self.transform:
            img = self.transform(img)

        # train.py와 동일한 형식으로 반환!
        return img, torch.FloatTensor(R), cont_labels, entry['front_image']

    def __len__(self):
        return len(self.data)

# ─────────────────────────────────────
# 2. 전처리 설정 (train.py와 동일)
# ─────────────────────────────────────
normalize = transforms.Normalize(
    mean=[0.485, 0.456, 0.406],
    std=[0.229, 0.224, 0.225])

transformations = transforms.Compose([
    transforms.Resize(240),
    transforms.CenterCrop(224),   # 테스트용이므로 CenterCrop
    transforms.ToTensor(),
    normalize
])

# ─────────────────────────────────────
# 3. 데이터셋 로드 & 확인
# ─────────────────────────────────────
json_path = '/content/drive/MyDrive/dl_프로젝트/final_260323_gt_annotations_all_포즈_50_80_제외.json'
dataset = TurtleNeckDataset(json_path, transform=transformations)

print(f"\n📊 데이터셋 크기: {len(dataset)}장")

# ─────────────────────────────────────
# 4. 샘플 1개 꺼내서 확인 (파이프라인 검증!)
# ─────────────────────────────────────
print("\n--- 샘플 데이터 확인 ---")
img, gt_R, cont_labels, filename = dataset[0]

print(f"  이미지 shape:    {img.shape}")          # [3, 224, 224] 이면 정상
print(f"  GT 회전행렬 shape: {gt_R.shape}")        # [3, 3] 이면 정상
print(f"  cont_labels:     {cont_labels}")         # [pitch°, yaw°, roll°]
print(f"  파일명:           {os.path.basename(filename)}")

print(f"\n  GT 회전행렬:")
print(f"  {gt_R}")

# ─────────────────────────────────────
# 5. 전체 데이터 분포 확인
# ─────────────────────────────────────
print("\n--- 전체 pitch 분포 ---")
for i in range(len(dataset)):
    _, _, labels, fname = dataset[i]
    pitch = labels[0].item()

    if pitch <= -10:
        bar = "◀◀◀"
        grade = "매우 정상"
    elif pitch <= 0:
        bar = "◀◀"
        grade = "정상"
    elif pitch <= 10:
        bar = "▶"
        grade = "경미 거북목"
    else:
        bar = "▶▶▶"
        grade = "심한 거북목"

    print(f"  [{i+1:2d}] pitch={pitch:6.1f}°  {bar:6s}  {grade:10s}  {os.path.basename(fname)}")

print("\n✅ Step 3 완료! 데이터셋이 TokenHPE에 들어갈 준비가 됐습니다.")
# ```

# ### 정상적으로 실행되면 이런 결과가 나옵니다
# ```
# ✅ 데이터셋 로드: 14장 (정면 paired)

# 📊 데이터셋 크기: 14장

# --- 샘플 데이터 확인 ---
#   이미지 shape:    torch.Size([3, 224, 224])     ← 이것이 나오면 성공!
#   GT 회전행렬 shape: torch.Size([3, 3])           ← 이것이 나오면 성공!
#   cont_labels:     tensor([-22.1, 0.0, 0.0])
#   파일명:           kki_front_10_010.jpg

# --- 전체 pitch 분포 ---
#   [ 1] pitch= -22.1°  ◀◀◀   매우 정상    kki_front_10_010.jpg
#   [ 2] pitch=  -9.9°  ◀◀     정상        kki_front_10_020.jpg
#   [ 3] pitch=  11.8°  ▶▶▶   심한 거북목   kki_front_10_051.jpg
#   ...

# ✅ Step 3 완료! 데이터셋이 TokenHPE에 들어갈 준비가 됐습니다.
````

````python
def Identity(self, cont_labels):
    """
    거북목 특화 9영역 (137장 데이터 기반, 최적 균형)

    pitch 경계: -14°, 0° → 46:44:47 (편차 3)
    yaw 경계: -15°, +15°

    pitch 기준:
      거북목:   pitch > 0°    (47장) — 머리가 앞으로 나온 상태
      경미:    -14° ~ 0°      (44장) — 정상~경계
      정상:     pitch ≤ -14°  (46장) — 확실히 좋은 자세
    """
    pitch = cont_labels[0]
    yaw = cont_labels[1]

    if pitch > 0:                               # 거북목 (47장)
        if yaw > 15:      flag = 0
        elif yaw < -15:   flag = 2
        else:             flag = 1
    elif pitch > -14:                           # 경미/경계 (44장)
        if yaw > 15:      flag = 3
        elif yaw < -15:   flag = 5
        else:             flag = 4
    else:                                       # 정상 (46장)
        if yaw > 15:      flag = 6
        elif yaw < -15:   flag = 8
        else:             flag = 7

    return flag
# ```

# ---

# ### 왜 바뀌었는가 (팀에게 설명용)
# ```
# 42장일 때:  경계 -10°, +5°  → 15:16:11
# 137장일 때: 경계 -14°,  0°  → 46:44:47

# 데이터가 3배 늘면서 "정상 자세" 쪽 데이터가 많이 추가됨
# → 경계를 왼쪽으로 이동해야 균형이 맞음
# → pitch 0°가 "거북목 시작" 기준이 되어 물리적으로도 자연스러움!
#    (0° = 머리가 어깨 바로 위 = 정상과 거북목의 자연 경계)
````

````python
# ============================================================
# loss.py ↔ datasets.py 연결 검증
# "Identity 함수가 우리 데이터에서 제대로 작동하는지 확인"
# ============================================================

# Identity 함수 (loss.py에서 가져옴)
def Identity_test(cont_labels):
    pitch = cont_labels[0].item()
    yaw = cont_labels[1].item()

    if pitch > 0:                               # 거북목 (변경: 5→0)
        if yaw > 15:      flag = 0
        elif yaw < -15:   flag = 2
        else:             flag = 1
    elif pitch > -14:                           # 경미 (변경: -10→-14)
        if yaw > 15:      flag = 3
        elif yaw < -15:   flag = 5
        else:             flag = 4
    else:                                       # 정상
        if yaw > 15:      flag = 6
        elif yaw < -15:   flag = 8
        else:             flag = 7
    return flag

# 전체 데이터에 대해 영역 배정 확인
region_names = {
    0: '거북목+왼쪽', 1: '거북목+정면', 2: '거북목+오른쪽',
    3: '경미+왼쪽',   4: '경미+정면',   5: '경미+오른쪽',
    6: '정상+왼쪽',   7: '정상+정면',   8: '정상+오른쪽'
}
region_count = {i: 0 for i in range(9)}

print("--- 42장 영역 배정 결과 ---\n")
for i in range(len(dataset)):
    img, gt_R, cont_labels, fname = dataset[i]
    region = Identity_test(cont_labels)
    region_count[region] += 1
    print(f"  [{i+1:2d}] pitch={cont_labels[0]:+6.1f}° "
          f"yaw={cont_labels[1]:+5.1f}° "
          f"→ 영역 {region} ({region_names[region]})"
          f"  {os.path.basename(fname)}")

print(f"\n--- 영역별 분포 ---")
print(f"{'영역':>4s}  {'이름':<14s}  {'수량':>4s}  {'비율':>6s}")
print("-" * 40)
for r in range(9):
    pct = region_count[r] / len(dataset) * 100
    bar = '█' * region_count[r]
    print(f"  {r}    {region_names[r]:<14s}  {region_count[r]:3d}장  "
          f"{pct:5.1f}%  {bar}")

# flag = -1 체크 (영역 미배정 오류)
unassigned = sum(1 for i in range(len(dataset))
                 if Identity_test(dataset[i][2]) == -1)
if unassigned > 0:
    print(f"\n⚠️ 영역 미배정: {unassigned}장 — Identity 경계 확인 필요!")
else:
    print(f"\n✅ 전체 {len(dataset)}장 영역 배정 완료! 미배정 0장 (오류 없음)")
# ```

# **정리하면, `-14°, 0°` 경계가 들어가야 할 곳이 3곳**입니다:
# ```
# 1. loss.py의 Identity 함수        ← 실제 학습에 사용
# 2. 검증 코드의 Identity_test 함수   ← 지금 수정하는 것
# 3. 결과 해석 시 등급 판정 기준      ← 나중에 eval.py
````

````python
# GT_data.json → GT_train.json / GT_test.json 분리
import json
import random

JSON_PATH  = '/content/drive/MyDrive/dl_프로젝트/GT_data.json'
TRAIN_PATH = '/content/drive/MyDrive/dl_프로젝트/GT_train.json'
TEST_PATH  = '/content/drive/MyDrive/dl_프로젝트/GT_test.json'

with open(JSON_PATH, 'r', encoding='utf-8') as f:
    data = json.load(f)

random.seed(42)
random.shuffle(data)

split = int(len(data) * 0.8)
train_data = data[:split]
test_data  = data[split:]

with open(TRAIN_PATH, 'w', encoding='utf-8') as f:
    json.dump(train_data, f, ensure_ascii=False, indent=2)

with open(TEST_PATH, 'w', encoding='utf-8') as f:
    json.dump(test_data, f, ensure_ascii=False, indent=2)

print(f'전체: {len(data)}장')
print(f'train: {len(train_data)}장 → GT_train.json')
print(f'test:  {len(test_data)}장  → GT_test.json')
# ```

# 실행하면 이렇게 나오면 성공입니다.
# ```
# 전체: 135장
# train: 108장 → GT_train.json
# test:  27장  → GT_test.json
````
